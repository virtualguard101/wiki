## Overview

[Jupyter Notebook](https://github.com/jupyter/notebook) 是一个开源的交互式计算环境，特别适合数据科学、机器学习、科学计算和教育领域。

其本质是一个 **JSON 格式**的“**交互式文档**”，前端（浏览器）负责展示，后端（kernel）负责执行代码，两者通过 WebSocket/HTTP 通信。

Jupyter Notebook 的主要作用在于，它能够让你在一个地方**边写代码，边看结果，边记笔记**。

### Main Features

- **交互式编程**：支持多种编程语言（Python、R、Julia、Scala等）
- **富文本支持**：Markdown、HTML、LaTeX 数学公式
- **可视化**：内联图表和图形显示
- **可重现性**：代码和结果保存在同一文档中
- **易于分享**：可以导出为多种格式（HTML、PDF、LaTeX等）

### Core Component

- **Notebook 文档**：包含代码单元格和文本单元格的 `.ipynb` 文件
- **Kernel**：执行代码的后端进程
- **Web 界面**：基于浏览器的交互式界面

### Main Uses

1. **数据科学**：数据清洗、探索性数据分析、建模
2. **机器学习**：算法开发、模型训练、结果可视化
3. **教育**：交互式教程、作业提交
4. **研究**：实验记录、结果展示
5. **原型开发**：快速验证想法和概念

## Installation

### Basic Installation

```bash
# 安装 Jupyter
pip install jupyter

# 启动 Jupyter Notebook
jupyter notebook

# 或使用 JupyterLab（更现代的界面）
pip install jupyterlab
jupyter lab
```

### Using Specific Kernel

```bash
# 安装特定内核
pip install ipykernel

# 将环境注册为 Jupyter 内核
python -m ipykernel install --user --name=myenv --display-name="Python (myenv)"
```


## Combination with MkDocs

可使用插件 [`mkdocs-jupyter`](https://github.com/danielfrg/mkdocs-jupyter) 使得 Jupyter Notebook 可以直接在基于[MkDocs](https://www.mkdocs.org/)的文档网站中渲染：

### Advantages

1. **无缝集成**：Notebook 可以直接在文档网站中渲染

2. **保持交互性**：支持代码执行和可视化

3. **版本控制**：`.ipynb` 文件可以很好地与 Git 配合使用

4. **统一管理**：所有文档和代码在一个地方

### Disadvantages

截至这篇笔记更新，该插件还没法很好地兼容[MkDocs Material](https://squidfunk.github.io/mkdocs-material/)主题中所支持的标记语法、页面布局、移动端页面适配等相对影响笔记查阅体验的功能。

### Configuration Example

#### Files to Include or Ignore

```yaml
plugins:
    - mkdocs-jupyter:
          include: ["*.ipynb"] # Default: ["*.py", "*.ipynb"]
          ignore: ["some-irrelevant-files/*.ipynb"]
```

#### Whether to Execute the Notebook before Conversion

```yaml
plugins:
    - mkdocs-jupyter:
          execute: true # Default: False
          # ignore the execution of some files (with glob matching)
          execute_ignore:
            - "my-secret-files/*.ipynb"
```

#### Configure Theme

```yaml
plugins:
    - mkdocs-jupyter:
          theme: dark

theme:
    name: material
    palette:
        scheme: slate
```

更多配置可参考插件的[Readme](https://github.com/danielfrg/mkdocs-jupyter?tab=readme-ov-file#configuration)

### Code Example

````python
# Example: Simple Data Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
data = {
    'x': np.random.randn(100),
    'y': np.random.randn(100)
}
df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['x'], df['y'], alpha=0.6)
plt.title('Random Data Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, alpha=0.3)
plt.show()

# Display data statistics
print("Data Statistics:")
print(df.describe(""))
````

## Best Practices

### Code Organization

- 保持单元格简洁，每个单元格只做一件事

- 使用有意义的变量名

- 添加注释说明复杂逻辑

### Documentation

- 使用 Markdown 单元格解释代码目的

- 添加数学公式和图表说明

- 保持叙述的连贯性

### Version Control

- 定期清理输出结果

- 使用 `.gitignore` 忽略临时文件

- 提交前检查代码质量

### Performance Optimization

- 避免在循环中重复计算

- 使用 `%timeit` 测量代码性能

- 考虑使用 `%%time` 测量整个单元格执行时间
