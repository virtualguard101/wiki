## 概述

[Jupyter Notebook](https://github.com/jupyter/notebook) 是一个开源的交互式计算环境，特别适合数据科学、机器学习、科学计算和教育领域。

其本质是一个 **JSON 格式**的“**交互式文档**”，前端（浏览器）负责展示，后端（kernel）负责执行代码，两者通过 WebSocket/HTTP 通信。

### 主要特点

- **交互式编程**：支持多种编程语言（Python、R、Julia、Scala等）
- **富文本支持**：Markdown、HTML、LaTeX 数学公式
- **可视化**：内联图表和图形显示
- **可重现性**：代码和结果保存在同一文档中
- **易于分享**：可以导出为多种格式（HTML、PDF、LaTeX等）

### 核心组件

- **Notebook 文档**：包含代码单元格和文本单元格的 `.ipynb` 文件
- **Kernel**：执行代码的后端进程
- **Web 界面**：基于浏览器的交互式界面

### 主要用途

1. **数据科学**：数据清洗、探索性数据分析、建模
2. **机器学习**：算法开发、模型训练、结果可视化
3. **教育**：交互式教程、作业提交
4. **研究**：实验记录、结果展示
5. **原型开发**：快速验证想法和概念

## 安装和使用

### 基本安装

```bash
# 安装 Jupyter
pip install jupyter

# 启动 Jupyter Notebook
jupyter notebook

# 或使用 JupyterLab（更现代的界面）
pip install jupyterlab
jupyter lab
```

### 在项目中使用

```bash
# 安装特定内核
pip install ipykernel

# 将环境注册为 Jupyter 内核
python -m ipykernel install --user --name=myenv --display-name="Python (myenv)"
```


## 与 MkDocs 集成

可使用 [mkdocs-jupyter插件](https://github.com/danielfrg/mkdocs-jupyter)使得 Jupyter Notebook 可以直接在基于[MkDocs](https://www.mkdocs.org/)的文档网站中渲染：

### 优势

1. **无缝集成**：Notebook 可以直接在文档网站中渲染

2. **保持交互性**：支持代码执行和可视化

3. **版本控制**：`.ipynb` 文件可以很好地与 Git 配合使用

4. **统一管理**：所有文档和代码在一个地方

### 配置示例

#### 要包含或忽略的文件

```yaml
plugins:
    - mkdocs-jupyter:
          include: ["*.ipynb"] # Default: ["*.py", "*.ipynb"]
          ignore: ["some-irrelevant-files/*.ipynb"]
```

#### 转换为文档前是否执行笔记本

```yaml
plugins:
    - mkdocs-jupyter:
          execute: true # Default: False
          # ignore the execution of some files (with glob matching)
          execute_ignore:
            - - "my-secret-files/*.ipynb"
```

#### 配置主题

```yaml
plugins:
    - mkdocs-jupyter:
          theme: dark

theme:
    name: material
    palette:
        scheme: slate
```

更多配置可参考插件的[Readme](https://github.com/danielfrg/mkdocs-jupyter?tab=readme-ov-file#configuration)

### 代码示例

````python
# Example: Simple Data Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
data = {
    'x': np.random.randn(100),
    'y': np.random.randn(100)
}
df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['x'], df['y'], alpha=0.6)
plt.title('Random Data Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, alpha=0.3)
plt.show()

# Display data statistics
print("Data Statistics:")
print(df.describe(""))
````

## 最佳实践

### 代码组织
- 保持单元格简洁，每个单元格只做一件事
- 使用有意义的变量名
- 添加注释说明复杂逻辑

### 文档化
- 使用 Markdown 单元格解释代码目的
- 添加数学公式和图表说明
- 保持叙述的连贯性

### 版本控制
- 定期清理输出结果
- 使用 `.gitignore` 忽略临时文件
- 提交前检查代码质量

### 性能优化
- 避免在循环中重复计算
- 使用 `%timeit` 测量代码性能
- 考虑使用 `%%time` 测量整个单元格执行时间
